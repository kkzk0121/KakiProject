```python
import random
import numpy as np
from pylab import plt

r = [(random.random() * 2 -1, random.random() * 2 - 1) for _ in range(500)]
#random.random()：0.0以上1.0未満の浮動小数点数
#リスト内包表記
#アンダースコア（_）は使用しない変数

r = np.array(r)

fig = plt.figure(figsize = (7, 7)) #新規のウィンドウ作成

ax = fig.add_subplot(1, 1, 1)
#add_subplot：描画キャンパスの分割
#引数：行方向の分割、列方向の分割、位置

circ = plt.Circle((0,0), radius = 1, edgecolor = 'g', facecolor = 'None', lw = 0.8) #円の生成
ax.add_patch(circ) #円の描画

circ = plt.Rectangle((-1,-1), 2, 2, edgecolor = 'b', alpha = 0.2) #正方形の生成
#alpha：透明度
ax.add_patch(circ) #正方形の描画

plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)

plt.style.use('seaborn') #図のスタイル設定

plt.plot(r[:, 0], r[:, 1], 'r.')

plt.show()
```

```python
distance = np.sqrt((r ** 2).sum(axis = 1))
print(distance[:5])
print((distance <= 1)[:5]) #点が半径１の円の中にあればTrue
print((distance <= 1).sum()) #半径１の円の中にある点の数
print(len(distance)) #点の合計

frac = ((distance <= 1).sum()/len(distance))
print(frac)

#πr^2 = 2r * 2r * frac
#π = 4 * frac
pi = 4 * frac
pi
```
